===

#1. 生成对抗网络基础
![images](images/01.png)

##1.1 GAN的组成
- 判别网络D:一个判别网络，判别一张图片是不是真实的，他的输入参数是x，x代表一张图片，输出D(x)代表x为真实图片的概率
- 生成网络G:一个生成图片的网络，它接收一个随机的噪声z，通过这个噪声生成图片，记作G(z)
- 损失Loss:判别是真实模型的概率

在训练过程中，生成网络G的目标就是尽量生成真是的图片去欺骗判别网络D，而D的目标就是尽量把G生成的图片和真实图片分别开来。这样G和D就构成了一个动态的"博弈过程"。在最理想的状态下，G可以生成足以"以假乱真"的图片G(z)。对于D来说，它难以判定G生成的图片究竟是不是真实的，因此D(G(z))=0.5

##1.2 GAN的数学基础
![images](images/02.png)
z和x是两个空间，我们的目标是将z变到x空间上
- 黑点组成的曲线，这是我们的目标曲线，标准分布
- 绿线：生成数据的分布
- 蓝线：判别函数(根据x大小判断是否真实的概率)

假设前提：判别模型D，生成模型G具有学习能力，能够收敛，我们的目标就是$min_Gmax_DV(D,G)=E_{x~p_{data}(x)}[logD(x)] + E_{z~p_z(z)}[log(1-D(G(z)))]$。对于D来说，就是正常的交叉熵损失，对于G来说，目标是$minLOG(1-D(G(z)))$

##1.3 GAN的特点

###1.3.1 优点
- 不需要大量label数据，loss来源于D判断
- 产生大量数据用于训练，接近无监督学习
- 可以和深度神经网络结合

###1.3.2 缺点
- 数据直接生成，没有推导过程
- 生成器、判别器需要配合共同训练难度较大
- 容易出现训练失败

#2. 深度GAN-DCGAN(Deep Convolutional GAN)
深度卷积网络与生成对抗网络的结合,DCGAN和GAN在原理上是一样的，它只是把上述的G和D换成了两个卷积神经网络，但不是直接替换就可以了，DCGAN对卷积神经网络的结构做了一些改变，以提高样本的质量和收敛速度.改变包括
- 取消所有的Pooling层：G网络中使用转置卷积进行上采样，D网络中用假如stride的卷积代替pooling
- 在D和G中均使用batch normalization
- 去掉FC层，是网络变为全卷积网络
- 激活函数：G网络使用ReLU作为激活函数，最后一层使用tanh；D网络使用LeakyReLU作为激活函数
