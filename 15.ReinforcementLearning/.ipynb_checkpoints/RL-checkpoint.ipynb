增强学习
===

#1 概述

##1.1 什么是增强学习
增强学习有叫做强化学习.![images](images/01.png)
- Agent:要学习的智能程序
- Policy:程序知道所处某状态后才去行为的策略(复杂情况DL，简单情况lookup table)
- Environment:智能程序交互外界环境，接受action产生状态变化，反悔reward，可以是真实世界、游戏模拟器、棋牌等等

##1.2 激励方程-reward function
它是某次行为直接产生的激励,用来定义增强学习的目的，学习的过程中，肯定是希望激励方程的回报越大越好.一个action产生一个立即的结果

##1.3 价值方程-value function
agent能够带来的长期累计回报。跟激励方程的区别就是：激励方程短期，价值方程长期

##1.4 环境模型-Model of the enviroment
环境模拟模型，模拟action作用之后环境返回的价值、激励、以及状态变化，模拟其中不需要，真实世界中有指导训练意义

##1.5 总结
增强学习的过程就是通过训练学到最有的state-action映射关系的过程，使agent得到最好的value/reward
